**Reasoning**:
I will construct a detailed prompt for the `llm_gemini` model, incorporating the instructions to generate a concise executive summary, a table of key metrics, and a dedicated section for risks and opportunities from the `report_content`. Then, I will invoke the `llm_gemini` model with this prompt and store the response in the `executive_summary` variable.



```python
from langchain_google_genai import ChatGoogleGenerativeAI
from google.colab import userdata

llm_gemini = ChatGoogleGenerativeAI(model="gemini-2.5-flash", google_api_key=userdata.get('Rup-ai-training'))

report_content = """This is a placeholder for the 10-page quarterly performance report content. It should contain details about financial performance, operational metrics, market conditions, and any identified risks or opportunities. For example, it might mention: \n- Revenue increased by 15% to $150M.\n- Net profit grew by 10% to $20M.\n- Customer acquisition cost decreased by 5%.\n- Market share expanded by 2%.\n- Key risk: increasing raw material costs due to supply chain disruptions.\n- Key opportunity: expansion into new international markets.\n- Operational efficiency improved by 8%.\n- Employee satisfaction remained high at 85%.\n- New product launch saw 20% higher than expected initial sales.\n- Regulatory changes pose a potential risk to future operations.\n- Strategic partnership discussions are underway, offering potential for significant growth."""

prompt_template = """You are a financial analyst. Your task is to create a concise executive summary (around 150 words) for the following quarterly performance report. The summary should include:
1. A brief overview of the quarter's performance.
2. A table of key performance metrics mentioned in the report.
3. A separate section outlining significant risks and opportunities identified in the report.

Quarterly Performance Report Content:
{report_content}

Executive Summary:"""

formatted_prompt = prompt_template.format(report_content=report_content)
executive_summary_response = llm_gemini.invoke(formatted_prompt)
executive_summary = executive_summary_response.content

print("Executive summary generated successfully.")
print("\n--- Executive Summary ---\n")
print(executive_summary)
```
