## Generate Concise Executive Summary with Metrics and Risks/Opportunities

### Subtask:
Utilize an LLM to process the provided report content and generate a concise executive summary (around 150 words) that includes a table of key metrics and a dedicated section outlining significant risks and opportunities mentioned in the report.


**Reasoning**:
First, I will create a placeholder for `report_content` as it was not provided in the notebook. This will allow the prompt to be constructed, and the user can replace this placeholder with their actual report content later.



```python
#Assignment:Client draft email
project_email_prompt_template = """You are a professional project manager drafting an email to a client. The purpose of this email is to provide an update on project progress, highlight key milestones achieved, outline upcoming actions, and politely request their feedback. Please ensure the email is professional, concise, and encourages engagement.

Draft the email below:
Subject: Project Update: [Project Name] Progress and Milestones - [Date]

Dear [Client Name],

We hope this email finds you well. We are writing to provide you with an update on the progress of the [Project Name] project, which remains on track for completion by [Project Deadline].

We are pleased to report that we have successfully achieved the following key milestones:
- [Milestone 1 Description and Date]
- [Milestone 2 Description and Date]
- [Milestone 3 Description and Date]

Our team is currently focused on [brief description of current phase/tasks].

Key Action Items for the Upcoming Period:
- [Action Item 1: Description, Owner, Due Date]
- [Action Item 2: Description, Owner, Due Date]
- [Action Item 3: Description, Owner, Due Date]

Your feedback is invaluable to us. Please let us know if you have any questions, comments, or require any adjustments at this stage. We are available to schedule a brief call at your convenience to discuss this update in more detail.

Thank you for your continued partnership.

Best regards,
[Your Name/Company Name]"""

project_email_response = llm_gemini.invoke(project_email_prompt_template)
project_email_draft = project_email_response.content

print("Project Update Email Draft generated successfully.")
print("\n--- Project Update Email Draft ---\n")
print(project_email_draft)
```

```python
# Assignment Executive Summary Generator
report_content = """This is a placeholder for the 10-page quarterly performance report content. It should contain details about financial performance, operational metrics, market conditions, and any identified risks or opportunities. For example, it might mention: \n- Revenue increased by 15% to $150M.\n- Net profit grew by 10% to $20M.\n- Customer acquisition cost decreased by 5%.\n- Market share expanded by 2%.\n- Key risk: increasing raw material costs due to supply chain disruptions.\n- Key opportunity: expansion into new international markets.\n- Operational efficiency improved by 8%.\n- Employee satisfaction remained high at 85%.\n- New product launch saw 20% higher than expected initial sales.\n- Regulatory changes pose a potential risk to future operations.\n- Strategic partnership discussions are underway, offering potential for significant growth."""
print("Placeholder 'report_content' created.")
```

**Reasoning**:
I will construct a detailed prompt for the `llm_gemini` model, incorporating the instructions to generate a concise executive summary, a table of key metrics, and a dedicated section for risks and opportunities from the `report_content`. Then, I will invoke the `llm_gemini` model with this prompt and store the response in the `executive_summary` variable.



```python
prompt_template = """You are a financial analyst. Your task is to create a concise executive summary (around 150 words) for the following quarterly performance report. The summary should include:
1. A brief overview of the quarter's performance.
2. A table of key performance metrics mentioned in the report.
3. A separate section outlining significant risks and opportunities identified in the report.

Quarterly Performance Report Content:
{report_content}

Executive Summary:"""

formatted_prompt = prompt_template.format(report_content=report_content)
executive_summary_response = llm_gemini.invoke(formatted_prompt)
executive_summary = executive_summary_response.content

print("Executive summary generated successfully.")
print("\n--- Executive Summary ---\n")
print(executive_summary)
```
