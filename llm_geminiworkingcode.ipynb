````python
# Check if Ollama process is running and try to curl its endpoint
import subprocess
import requests

try:
    # Check Ollama processes
    print("--- Checking Ollama processes ---")
    subprocess.run(["ollama", "ps"], check=True)
    print("\n--- Attempting to connect to Ollama API ---")
    response = requests.get("http://localhost:11434")
    response.raise_for_status() # Raise an exception for HTTP errors
    print(f"Successfully connected to Ollama API. Status code: {response.status_code}")
    print("Response: " + response.text)
except subprocess.CalledProcessError as e:
    print(f"Error checking Ollama processes: {e}")
except requests.exceptions.ConnectionError as e:
    print(f"Connection to Ollama API refused: {e}")
except requests.exceptions.RequestException as e:
    print(f"An unexpected error occurred while connecting to Ollama API: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
````

### 2. Run Ollama Server in Background

````python
# Start Ollama in the background
import subprocess
import time

# Set OLLAMA_HOST environment variable
%env OLLAMA_HOST=0.0.0.0:11434

# Start Ollama server as a background process
process = subprocess.Popen(["ollama", "serve"], stdout=subprocess.PIPE, stderr=subprocess.PIPE)

# Give Ollama some time to start up
time.sleep(5)
print("Ollama server started in the background.")
````
